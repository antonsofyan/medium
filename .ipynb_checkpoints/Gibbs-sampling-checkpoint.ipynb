```python
import numpy as np
import scipy as sc
from scipy.stats import invgamma,invwishart, norm,multivariate_normal
import math
import matplotlib.pyplot as plt
%matplotlib inline
```

# Gibbs sampling for the 1D Gaussian, infering mean and variance

## Generate data and define the distributions

```python
#Generate data from the normal distribution
data = np.random.normal(loc=2, scale=5,size=1000)
# Define the conditional posteriors for mean and variance
# The mean distribution follows the noraml distribution conditional on the data and the variance
normal_dist = lambda data, var: np.random.normal(loc = np.mean(data), scale = var/len(data))
# The variance distribution is the inverse gamma distribution conditional on the data and the mean 
# Note in scipy.stats scale parameter is equivalent to beta in the wikipedia convention
inverse_gamma = lambda data,mu: invgamma.rvs(a=len(data)/2,scale=sum((data-mu)**2)/2) 
```

## Run Gibbs sampler

```python
# Lists to hold the sampled values
mu_list = list()
std_list = list()
# Starting point for the mean
mu = 0
mu_list.append(mu)
# How many iterations to sample
number_of_iterations = 2000
# Iteratively sample
for ii in range(0,number_of_iterations):
    # Sample the variance conditional on the previous mean and data
    var = inverse_gamma(data,mu)
    # Sample the mean conditional on the previous mean and data
    mu = normal_dist(data,var)
    # Record the values
    mu_list.append(mu)
    std_list.append(math.sqrt(var))
```

## Print and plot the results

```python
print(f'Emperical mean: {round(np.mean(data),3)}, estimate from the Gibbs sampler: {round(mu_list[-1],3)}')
print(f'Emperical std: {round(np.std(data),3)}, estimate from the Gibbs sampler: {round(std_list[-1],3)}')
```

```python
#Plot variance over time
plt.figure()
plt.plot(std_list)
plt.xlabel('Iteration', fontsize=12)
plt.ylabel('Standard deviation', fontsize=12)
plt.savefig('var_1d')

#Plot mean over time
plt.figure()
plt.plot(mu_list)
plt.xlabel('Iteration', fontsize=12)
plt.ylabel('Mean', fontsize=12)
plt.savefig('mean_1d')

#Plot distribution and data
plt.figure()
xx = np.linspace(mu_list[-1]-3*std_list[-1],mu_list[-1]+3*std_list[-1],100)
yy = norm.pdf(xx, loc=mu_list[-1], scale=std_list[-1])
plt.hist(data,color='g', bins=100, density=True, alpha = 0.2, label = 'Data')
plt.plot(xx, yy, color = 'b', label = 'Estimated distribution')
plt.xlabel('Value', fontsize=12)
plt.ylabel('Probability', fontsize=12)
plt.legend(fontsize=12)
plt.axvline(x=mu_list[-1],color='r')
ax = plt.gca()
plt.savefig('1d_distribution')
plt.show()
```

# Gibbs sampling for the 2D Gaussian, infering mean and covariance matrix

## Generate data and define the distributions

```python
# Generate data
data2d = np.random.multivariate_normal(mean = np.array([1,3]), cov = np.array([[2,0.9],[0.9,10]]),size = 10000)

# Define the posterior for the mean, conditional on the covariance matrix and data
mult_normal_dist = lambda data2d, cov: np.random.multivariate_normal(mean = (np.mean(data2d, axis=0)), cov =cov/np.size(data2d,0))

# Define the posterior for the covariance matrix, conditional on the mean and data
inv_wishart_dist = lambda data2d, mean: invwishart.rvs(df = np.size(data2d,0), scale = np.matmul(np.transpose((data2d - mean)),(data2d - mean)))
```

## Run Gibbs sampler

```python

# Starting point for the mean
mean = np.array([-1,1])
# How many iterations to sample
number_of_iterations = 1000
mean_1 = list()
mean_2 = list()
std_1 = list()
std_2 = list()
cov_l = list()
mean_1.append(mean[0])
mean_2.append(mean[1])
# Iteratively sample
for ii in range(0,number_of_iterations):
    # Sample the variance conditional on the previous mean and data
    cov = inv_wishart_dist(data2d, mean)
    std_1.append(cov[0,0])
    std_2.append(cov[1,1])
    cov_l.append(cov[1,0])
    # Sample the mean conditional on the previous mean and data
    mean = mult_normal_dist(data2d,cov)
    mean_1.append(mean[0])
    mean_2.append(mean[1])
```

## Print and plot the results

```python
print('True mean: ', np.mean(data2d,0), 'Estimated mean:', mean)
print('True cov: \n', np.cov(data2d.T), '\nEstimated cov:\n', cov)
```

```python
# Plot covariance over time
plt.figure()
plt.plot(cov_l,'b')
plt.xlabel('Iteration', fontsize=12)
plt.ylabel('Covariance', fontsize=12)
plt.savefig('covariance.png')


# Plot variances over time
plt.figure()
plt.plot(std_1,'b', label='std1')
plt.plot(std_2,'g', label='std2')
plt.xlabel('Iteration', fontsize=12)
plt.ylabel('Standard deviation', fontsize=12)
plt.legend(fontsize=12)
plt.savefig('var2d.png')

# Plot mean over time
plt.figure()
plt.plot(mean_1,'b', label='mean1')
plt.plot(mean_2,'g', label='mean2')
plt.xlabel('Iteration', fontsize=12)
plt.ylabel('Mean', fontsize=12)
plt.legend(fontsize=12)
plt.savefig('mean2d.png')

# Plot the whole distribution and the data
plt.figure()
x = np.linspace(-10,10,500)
y = np.linspace(-10,15,500)
X,Y = np.meshgrid(x,y)
pos = np.array([X.flatten(),Y.flatten()]).T
rv = multivariate_normal(mean, cov)
rv = multivariate_normal(mean, cov)

ax = plt.gca()

ax.contour(X,Y,rv.pdf(pos).reshape(500,500))
ax.autoscale(False)
plt.scatter(data2d[:,0],data2d[:,1],color = 'g',zorder=1, alpha = 0.2)
plt.xlabel('Data x', fontsize= 12)
plt.ylabel('Data y', fontsize= 12)
plt.savefig('2d_normal')
```
